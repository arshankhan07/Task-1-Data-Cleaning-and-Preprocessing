```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
```

```python
original_rows, original_cols = df.shape
```

```python
missing_before = df.isnull().sum().sum()
```

```python
duplicates_before = df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
```

```python
df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
    .str.replace("-", "_")
)
```

```python
df['customerid'] = df['customerid'].astype(int)
df['age'] = df['age'].astype(int)
df['annual_income_k$'] = df['annual_income_k$'].astype(int)
df['spending_score_1_100'] = df['spending_score_1_100'].astype(int)
```

```python
final_rows, final_cols = df.shape
```

```python
clean_path = "/content/cleaned_data.csv"
df.to_csv(clean_path, index=False)

original_rows, original_cols, missing_before, duplicates_before, final_rows, final_cols
```
